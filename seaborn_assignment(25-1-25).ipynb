```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot  as plt
data=pd.read_csv("train.csv")
data
```

```python
sns.barplot(data=data,x="Pclass",y="Survived")
plt.title("survival rate by passenger")
plt.xlabel("passenger class")
plt.ylabel("survival")
plt.show()
```

```python
sns.scatterplot(data=data,x="Age",y="Fare",hue="Sex",s=80,palette="magma",alpha=0.6)
plt.title("Age vs Fare")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.show()               
```

```python
pair_class=data[["Age","Fare","Survived","Pclass"]]
sns.pairplot(data=pair_class,hue="Survived",palette="magma")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
numeric_col=data.select_dtypes(include=['number'])
corelation_matrix=numeric_col.corr()
sns.heatmap(corelation_matrix,annot=True,cmap="magma",fmt=".1f")
plt.title("heatmap of corelation")
plt.show()
```
