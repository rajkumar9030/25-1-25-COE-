```python
import seaborn as sns
import matplotlib.pyplot as plt
tips=sns.load_dataset("tips")
tips.head()
```

```python
sns.histplot(data=tips,x="total_bill",kde=True,bins=20,color="skyblue")
plt.title("histogram in seaborn")

plt.show()
```

```python
sns.kdeplot(data=tips,x="total_bill",fill=True,color="green")
plt.title("kde visua;ization")
plt.show()
```

```python
sns.barplot(data=tips,x="day",y="total_bill",palette= 'viridis')
plt.title("avg total bill per day")
plt.show()
```

```python
sns.boxplot(data=tips,x="day",y="total_bill",hue="sex",palette= 'magma')
plt.title(" total bill per day by sex")
plt.show()
```

```python
sns.scatterplot(data=tips,x="total_bill",y="tip",hue="day",palette= 'inferno',style="time")
plt.title(" scatter plot total bill vs tip")
plt.show()
```

```python
sns.lineplot(data=tips,x="size",y="tip",marker='*',color="red")
plt.title(" line plot size vs tip")
plt.show()
```

```python
sns.pairplot(data=tips,hue="sex",palette="magma")

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
numeric_col=tips.select_dtypes(include=['int64','float64'])
corelation_matrix=numeric_col.corr()
sns.heatmap(corelation_matrix,annot=True,cmap="magma",fmt=".2f")
plt.title("heatmap of corelation")
plt.show()
```
